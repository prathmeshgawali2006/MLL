```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_sal= pd.read_csv(r"C:\Users\asus\Downloads\Salary_Data.csv")

print(df_sal.head())


print(df_sal.describe())

plt.title('Salary Distribution Plot')
sns.histplot(df_sal['Salary'], kde=True)  
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.show()

plt.title('Salary Distribution of Salary')

sns.regplot(x='YearsExperience',y='Salary',data=df_sal)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np


df_sal = pd.read_csv(r"C:\Users\asus\Downloads\Salary_Data.csv")


X= df_sal[['YearsExperience']]
y= df_sal['Salary']

model = LinearRegression()
model.fit(X, y)


y_pred = model.predict(X)


slope = model.coef_[0]
intercept = model.intercept_
rmse = np.sqrt(mean_squared_error(y, y_pred))


print(f"Slope (Coefficient): {slope}")
print(f"Intercept: {intercept}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
    

plt.figure(figsize=(8, 6))
sns.scatterplot(x='YearsExperience', y='Salary', data=df_sal, label='Actual')
plt.plot(df_sal['YearsExperience'], y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Salary vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
